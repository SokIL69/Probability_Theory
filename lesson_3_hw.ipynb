–£—Ä–æ–∫ 3.

2. –í –ø–µ—Ä–≤–æ–º —è—â–∏–∫–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è 8 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 - –±–µ–ª—ã–µ. –í–æ –≤—Ç–æ—Ä–æ–º —è—â–∏–∫–µ - 12 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 –±–µ–ª—ã—Ö. –ò–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ –≤—ã—Ç–∞—Å–∫–∏–≤–∞—é—Ç —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –¥–≤–∞ –º—è—á–∞, –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ - 4. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ 3 –º—è—á–∞ –±–µ–ª—ã–µ?
3. –ù–∞ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ –±–∏–∞—Ç–ª–æ–Ω—É –æ–¥–∏–Ω –∏–∑ —Ç—Ä–µ—Ö —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ —Å—Ç—Ä–µ–ª—è–µ—Ç –∏ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –º–∏—à–µ–Ω—å. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞ —Ä–∞–≤–Ω–∞ 0.9, –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ ‚Äî 0.8, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–≥–æ ‚Äî 0.6. –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω: a). –ø–µ—Ä–≤—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º –±). –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º –≤). —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º.
4. –í —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç—ã A –∏ B –ø–æ—Å—Ç—É–ø–∏–ª–æ —Ä–∞–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∞ –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç C —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ—Å—Ç—É–ø–∏–ª–æ —Å—Ç–æ–ª—å–∫–æ –∂–µ, —Å–∫–æ–ª—å–∫–æ –Ω–∞ A –∏ B –≤–º–µ—Å—Ç–µ. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ A —Å–¥–∞—Å—Ç –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é, —Ä–∞–≤–Ω–∞ 0.8. –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ B —ç—Ç–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 0.7, –∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ C - 0.9. –°—Ç—É–¥–µ–Ω—Ç —Å–¥–∞–ª –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–Ω —É—á–∏—Ç—Å—è: a). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A –±). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B –≤). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C?
5. –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Ç—Ä–µ—Ö –¥–µ—Ç–∞–ª–µ–π. –î–ª—è –ø–µ—Ä–≤–æ–π –¥–µ—Ç–∞–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–π—Ç–∏ –∏–∑ —Å—Ç—Ä–æ—è –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü —Ä–∞–≤–Ω–∞ 0.1, –¥–ª—è –≤—Ç–æ—Ä–æ–π - 0.2, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–π - 0.25. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –≤—ã–π–¥—É—Ç –∏–∑ —Å—Ç—Ä–æ—è: –∞). –≤—Å–µ –¥–µ—Ç–∞–ª–∏ –±). —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏ –≤). —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å –≥). –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π?

```python
from math import factorial
import pandas as pd
import numpy as np
```

–ß–∏—Å–ª–æ —Å–æ—á–µ—Ç–∞–Ω–∏–π

$$C_{n}^k =\frac{n!}{k!(n - k)!}$$ 

```python
def combinations(n, k):
    # –ß–∏—Å–ª–æ —Å–æ—á–µ—Ç–∞–Ω–∏–π –∏–∑  ùëõ  —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø–æ  ùëò  —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –∫–∞–∂–¥–æ–º (–≤ —Å–æ—á–µ—Ç–∞–Ω–∏—è—Ö –ø–æ—Ä—è–¥–æ–∫ –Ω–µ –≤–∞–∂–µ–Ω):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

## –ó–∞–¥–∞—á–∞ 1

–î–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –∏–∑ –≤—ã–±–æ—Ä–∫–∏ –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. 

–ü–æ—Å—á–∏—Ç–∞—Ç—å (–∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –Ω–∞–ø–æ–¥–æ–±–∏–µ std, var, mean)
- —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ,
- —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ,
- —Å–º–µ—â–µ–Ω–Ω—É—é –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.

```python
X = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

N = len(X)  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ
X_cp = np.sum(X)/N  # –°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–∏—Ç–∏—á–µ—Å–∫–æ–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏
```

```python
# 1 –≤–∞—Ä–∏–∞–Ω—Ç
sum = 0
for x in X:
    sum += (x - X_cp)**2

print(sum)

# 2-–π –≤–∞—Ä–∏–∞–Ω—Ç
arr = np.array(X) - X_cp
sum = np.sum(arr**2)
print(sum)

# 3-–π –≤–∞—Ä–∏–∞–Ω—Ç
sum = np.dot(arr, arr)
print(sum)

S_cm = sum/N  # C–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.
S_ncm = sum/(N-1)  # –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.
```

```python
print(f'–°—É–º–º–∞ –∑–Ω–∞—á–µ–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ = {np.sum(X)}')
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ = {N}\n")

print("–û—Ç–≤–µ—Ç:\n")
print(f"1. –°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–∏—Ç–∏—á–µ—Å–∫–æ–µ –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {X_cp}")
print(f"2. –°–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {S_cm}")
print(f"3. –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {S_ncm}")
print(f"4. –°–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è (–°–¢–û) –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {np.sqrt(S_cm)}")
print(f"5. –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è (–°–¢–û) –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {np.sqrt(S_ncm)}")
```

```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞
print('–ü—Ä–æ–≤–µ—Ä–∫–∞:\n')
print(f"1. –°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑/–ø = {np.mean(X)}")
print(f"2. –°–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {np.var(X)}")
print(f"3. –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {np.var(X, ddof=1)}")
print(f"4. –°–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è (–°–¢–û) –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {np.std(X)}")
print(f"5. –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è (–°–¢–û) –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ = {np.std(X, ddof=1)}")
```

## –ó–∞–¥–∞—á–∞ 2

–í –ø–µ—Ä–≤–æ–º —è—â–∏–∫–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è 8 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 - –±–µ–ª—ã–µ. –í–æ –≤—Ç–æ—Ä–æ–º —è—â–∏–∫–µ - 12 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 –±–µ–ª—ã—Ö.

–ò–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ –≤—ã—Ç–∞—Å–∫–∏–≤–∞—é—Ç —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –¥–≤–∞ –º—è—á–∞, –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ - 4.

–ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ 3 –º—è—á–∞ –±–µ–ª—ã–µ?

A11- –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ –≤—ã–Ω—É–ª–∏ –¥–≤–∞ –±–µ–ª—ã—Ö –º—è—á–∞

–ê12 - –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ —è—â–∏–∫–∞ –≤—ã–Ω—É–ª–∏ –æ–¥–∏–Ω –±–µ–ª—ã–π –º—è—á

–ê21 - –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ –≤—ã–Ω—É–ª–∏ –æ–¥–∏–Ω –±–µ–ª—ã–π –º—è—á

–ê22 - –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ —è—â–∏–∫–∞ –≤—ã–Ω—É–ª–∏ –¥–≤–∞ –±–µ–ª—ã—Ö –º—è—á–∞

–ê31 - –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ –≤—ã–Ω—É–ª–∏ —Ç–æ–ª—å–∫–æ —á–µ—Ä–Ω—ã–µ –º—è—á–∏

–ê32 - –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ —è—â–∏–∫–∞ –≤—ã–Ω—É–ª–∏ —Ç—Ä–∏ –±–µ–ª—ã–π –º—è—á–∏

–ê: –ê11 –∏ –ê12 –∏–ª–∏ –ê21 –∏ –ê22 –∏–ª–∏ –ê31 –∏ –ê32 (–≤—ã–Ω—É–ª–∏ —Ä–æ–≤–Ω–æ —Ç—Ä–∏ –±–µ–ª—ã—Ö –º—è—á–∞)

$P(A11) =  \frac{C_{5}^{2}}{C_{8}^{2} }$

$P(A12) =\frac{C_{5}^{1}}{C_{12}^{4} }$

```python
P_A11 = combinations(5, 2)/combinations(8, 2)
print(f"P_A11 = {P_A11*100:.1f} % ({P_A11})")
P_A12 = combinations(5, 1)/combinations(12, 4)
print(f"P_A2 = {P_A12*100:.1f} %  ({P_A12})")
```

$P(A21) =  \frac{C_{5}^{1}}{C_{8}^{2} }$

$P(A22) =\frac{C_{5}^{2}}{C_{12}^{4} }$

```python
P_A21 = combinations(5, 1)/combinations(8, 2)
print(f"P_A21 = {P_A21*100:.1f} %  ({P_A21})")
P_A22 = combinations(5, 2)/combinations(12, 4)
print(f"P_A22 = {P_A22*100:.1f} %  ({P_A22})")
```

$P(A31) =  \frac{C_{3}^{2}}{C_{8}^{2} }$

$P(A32) =\frac{C_{5}^{3}}{C_{12}^{4} }$

```python
P_A31 = combinations(3, 2)/combinations(8, 2)
print(f"P_A31 = {P_A31*100:.1f} %  ({P_A31})")
P_A32 = combinations(5, 3)/combinations(12, 4)
print(f"P_A32 = {P_A32*100:.1f} %  ({P_A32})")
```

```python
PA = P_A11*P_A12 + P_A21*P_A22 + P_A31*P_A32
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã–Ω—É—Ç—å —Ä–æ–≤–Ω–æ —Ç—Ä–∏ –±–µ–ª—ã—Ö –º—è—á–∞ = {PA*100 :.3f} %  ({PA})')
```

**–û—Ç–≤–µ—Ç**: –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã–Ω—É—Ç —Ä–æ–≤–Ω–æ –¥–≤–∞ –±–µ–ª—ã—Ö –º—è—á–∞ P(A)= 0.94%

## –ó–∞–¥–∞—á–∞ 3

–ù–∞ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ –±–∏–∞—Ç–ª–æ–Ω—É –æ–¥–∏–Ω –∏–∑ —Ç—Ä–µ—Ö —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ —Å—Ç—Ä–µ–ª—è–µ—Ç –∏ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –º–∏—à–µ–Ω—å.

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è:
- –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞ —Ä–∞–≤–Ω–∞  P(A|B1)=0.9,
- –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ ‚Äî  P(A|B2)=0.8,
- –¥–ª—è —Ç—Ä–µ—Ç—å–µ–≥–æ ‚Äî P(A|B3)=0.6.

–ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω:

3. —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º.

- –°–æ–±—ã—Ç–∏–µ –ê - –ø–æ–ø–∞–¥–∞–Ω–∏–µ –≤ –º–∏—à–µ–Ω—å.
- –°–æ–±—ã—Ç–∏–µ B1 - –≤—ã—Å—Ç—Ä–µ–ª —Å–¥–µ–ª–∞–Ω 1-–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º. $P(B1)=\frac{1}{3}$
- –°–æ–±—ã—Ç–∏–µ B2 - –≤—ã—Å—Ç—Ä–µ–ª —Å–¥–µ–ª–∞–Ω 2-–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º. $P(B2)=\frac{1}{3}$
- –°–æ–±—ã—Ç–∏–µ B3 - –≤—ã—Å—Ç—Ä–µ–ª —Å–¥–µ–ª–∞–Ω 3-–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º. $P(B3)=\frac{1}{3}$

$$P(Bi|A) = \frac{P(Bi)P(A|Bi)}{P(A)}$$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –º–∏—à–µ–Ω—å –Ω–∞—Ö–æ–¥–∏–º –ø–æ —Ñ–æ—Ä–º—É–ª–µ –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:
$$P(A) = \frac{1}{3}P(A|B1)+\frac{1}{3}P(A|B2)+\frac{1}{3}P(A|B3)$$

–£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Ç—Ä–µ—Ö –¥–µ—Ç–∞–ª–µ–π.

- –¥–ª—è –≤—Ç–æ—Ä–æ–π - 0.2,
- –¥–ª—è —Ç—Ä–µ—Ç—å–µ–π - 0.25.

–ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –≤—ã–π–¥—É—Ç –∏–∑ —Å—Ç—Ä–æ—è:
- (–∞) –≤—Å–µ –¥–µ—Ç–∞–ª–∏?
- (–±) —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏?
- (–≤) —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å?
- (–≥) –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π?

- –°–æ–±—ã—Ç–∏–µ B1 - –∏–∑ —Å—Ç—Ä–æ—è –≤—ã—à–ª–∞ 1-—è –¥–µ—Ç–∞–ª—å. $P(B_1)=0.1$
- –°–æ–±—ã—Ç–∏–µ B2 - –∏–∑ —Å—Ç—Ä–æ—è –≤—ã—à–ª–∞ 2-—è –¥–µ—Ç–∞–ª—å. $P(B_2)=0.2$
- –°–æ–±—ã—Ç–∏–µ B3 - –∏–∑ —Å—Ç—Ä–æ—è –≤—ã—à–ª–∞ 3-—è –¥–µ—Ç–∞–ª—å. $P(B_3)=0.25$

**–ê1**: –Ω–µ B1 –∏ –Ω–µ B2 –∏ –Ω–µ B3 - **–Ω–∏ –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å –Ω–µ –≤—ã—à–ª–∞ –∏–∑ —Å—Ç—Ä–æ—è**

$$P(A_1) = (1-ùëÉ(ùêµ_1))(1-ùëÉ(ùêµ_2))(1-ùëÉ(ùêµ_3))=0.9*0.8*0.75=0.54=54 \% $$

**–ê2**: (B1 –∏ –Ω–µ B2 –∏ –Ω–µ B3) –∏–ª–∏ (B2 –∏ –Ω–µ B1 –∏ –Ω–µ B3 ) –∏–ª–∏ (B3 –∏ –Ω–µ B1 –∏ –Ω–µ B2) - **–∏–∑ —Å—Ç—Ä–æ—è –≤—ã—à–ª–∞ —Ç–æ–ª—å–∫–æ –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å**

$$P(A_2) = ùëÉ(ùêµ_1)(1 - P(B_2))(1 - P(B_3)) + ùëÉ(ùêµ_2)(1 - P(B_1))(1 - P(B_3)) + ùëÉ(ùêµ_3)(1 - P(B_1))(1 - P(B_2)) = 0.1*0.8*0.75 + 0.2*0.9*0.75 + 0.25*0.9*0.8 = 0.375 = 37.5\%$$

**–ê3**: (B1 –∏ B2 –∏ –Ω–µ B3) –∏–ª–∏ (B1 –∏ B3 –∏ –Ω–µ B2) –∏–ª–∏ (B2 –∏ B3 –∏ –Ω–µ B1) - **–∏–∑ —Å—Ç—Ä–æ—è –≤—ã—à–ª–∏ –¥–≤–µ –¥–µ—Ç–∞–ª–∏**
